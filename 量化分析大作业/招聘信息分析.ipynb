```python
'''
数据集job_data.xlsx为互联网上收集的不同职位需求数据，对该数据进行分析并对每个小题目进行可视化。
1.	国内不同地点招聘人数总数
2.	不同部门的职位招聘人数总数
3.	不同类别岗位的学历要求分析
4.	不同类别岗位的平均工作年限分析
5.	国内不同地点工作类型分析
6.	不同月份发布的职位需求人数分析
'''
```

```python
'''
国内不同地点招聘人数总数
'''
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 读取数据
dates = pd.read_excel("job_data.xlsx")
dit = {}
for date in dates["工作地点"]:
    if date not in dit.keys():
        dit[date] = 1
    else:
        dit[date] = dit[date] + 1
name = list(dit.keys())
num = list(dit.values())
plt.title("国内不同地点招聘人数")
plt.bar(name,num)
plt.xlabel("地点")
plt.ylabel("人数")
plt.xticks(rotation=90)
plt.show()
```

```python
'''
不同部门的职位招聘人数总数
'''
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 读取数据
dates = pd.read_excel("job_data.xlsx")
dit = {}
for date in dates["所属部门"]:
    if date not in dit.keys():
        dit[date] = 1
    else:
        dit[date] += 1
part = list(dit.keys())
num = list(dit.values())
plt.title("不同部门的职位招聘人数")
plt.bar(part[0:20],num[0:20])
plt.xlabel("部门")
plt.ylabel("人数")
plt.xticks(rotation=90)
plt.show()
```

```python
'''
不同类别岗位的学历要求分析
'''
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 读取数据
dates = pd.read_excel("job_data.xlsx")
dit = {}
for parting in dates["职位类别"]:
    dit[parting] = []
for n in range(len(dates["学历要求"])):
    dit[dates["职位类别"][n]].append(dates["学历要求"][n])
for key,value in dit.items():
    dit1 = {}
    for keys in value:
        if keys not in dit1.keys():
            dit1[keys] = 1
        else:
            dit1[keys] += 1
    dit[key] = dit1
for key,value in dit.items():
    education = list(value.keys())
    num = list(value.values())
    all = 0
    for ave in num:
        all += ave
    num = [x/all for x in num]
    plt.pie(num,labels=education,autopct='%1.2f%%')
    plt.title(key)
    plt.show()
```

```python
'''
不同类别岗位的平均工作年限分析
'''
dit = {}
for parting in dates["职位类别"]:
    dit[parting] = []
for n in range(len(dates["工作年限"])):
    dit[dates["职位类别"][n]].append(dates["工作年限"][n])
for key,value in dit.items():
    dit1 = {}
    for keys in value:
        if keys not in dit1.keys():
            dit1[keys] = 1
        else:
            dit1[keys] += 1
    dit[key] = dit1
for key,value in dit.items():
    education = list(value.keys())
    num = list(value.values())
    all = 0
    for ave in num:
        all += ave
    num = [x/all for x in num]
    plt.pie(num,labels=education,autopct='%1.2f%%')
    plt.title(key)
    plt.show()
```
