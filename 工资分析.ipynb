```python
import matplotlib as mpl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame,Series

%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']  #
plt.rcParams['axes.unicode_minus'] = False  #
filename = "Baltimore_City_Employee_Salaries_FY2016.csv"
data = pd.read_csv(filename)
colors = ["red","green","blue","black","grey","purple"]
```

```python
#1. 
months = {}
for month in data["HireDate"]:
    if month[0:2] in months.keys():
        months[month[0:2]] += 1
    else:
        months[month[0:2]] = 1
months = dict(sorted(months.items(), key=lambda x: x[1], reverse=True))
month = list(months.keys())
counts = list(months.values())
plt.bar(month,counts,color=colors)
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```

```python
""""""

data.sort_values("AnnualSalary",inplace=True,ascending=False)
plt.bar(data.iloc[5::-1,1],data.iloc[5::-1,5],color=colors)
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```

```python
count = {}
for Job in data["JobTitle"]:
    if Job in count.keys():
        count[Job] += 1
    else:
        count[Job] = 1
sort_count = sorted(count.items(),key=lambda x:x[1],reverse=True)
# print(sort_count[:5])
# data.sort_values("AnnualSalary",inplace=True,ascending=False)
key_name=[]
values = []
for i in range(5):
    key_name.append(sort_count[i][0])
    values.append(sort_count[i][1])
plt.bar(key_name[-1::-1],values[-1::-1],color=colors)
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.show()
```
